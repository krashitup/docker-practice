```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col,udf
from pyspark.sql.types import StringType
# Initialize
spark = SparkSession.builder.appName("SparkCompleteNotes").getOrCreate()
# Create Base DataFrame
data = [
    (1, "Alice", "Engineering", 75000, 15),
    (2, "Bob", "Marketing", 60000, 30),
    (3, "Charlie", "Engineering", 80000, 15),
    (4, "David", "Sales", 65000, 28),
    (5, "Eve", "Marketing", 70000, 12)
]
columns = ["Id", "Name", "Department", "Salary", "Age"]
df = spark.createDataFrame(data, columns);
```

```python
df_filter = df.filter(col("Salary") > 70000)
for element in df_filter.select():
    print(element)
```

```python
def driving_eligibility(age):
    if age >= 18:
        return "Can Drive"
    elif age >= 13:
        return "Learning License"
    else:
        return "Cannot Drive"
```

```python
driving_udf = udf(driving_eligibility, StringType())
```

```python
result_df = df.withColumn(
    "Driving_Status",
    driving_udf(df.Age)
)

```
